```python
from sklearn.cluster import DBSCAN
import numpy as np

# Define the dataset
X = np.array([[1, 1], [2, 1], [3, 1], [5, 1], [6, 1], [7, 1], [4, 4], [5, 4], [6, 4], [5, 5], 
              [2, 2], [2, 3], [3, 2], [3, 3], [8, 8], [8, 9], [9, 8], [9, 9], [10, 10], [11, 10], 
              [12, 10], [10, 11], [11, 11], [12, 11], [10, 12], [11, 12], [12, 12], [15, 15], [16, 15], 
              [17, 15], [15, 16], [16, 16], [17, 16], [15, 17], [16, 17], [17, 17], [20, 5], [21, 5], 
              [22, 5], [20, 6], [21, 6], [22, 6], [20, 7], [21, 7], [22, 7], [18, 10], [19, 10], [20, 10], 
              [18, 11], [19, 11], [20, 11], [18, 12], [19, 12], [20, 12], [15, 20], [16, 20], [17, 20], 
              [15, 21], [16, 21], [17, 21], [15, 22], [16, 22], [17, 22],
              [21, 9], [22, 9], [23, 9], [21, 10], [22, 10], [23, 10], [21, 11], [22, 11], [23, 11], 
              [10, 5], [11, 5], [12, 5], [10, 6], [11, 6], [12, 6], [10, 7], [11, 7], [12, 7]])

# Instantiate DBSCAN object
dbscan = DBSCAN(eps=3, min_samples=5)

# Fit the DBSCAN object to the dataset
dbscan.fit(X)

# Get the labels assigned to each instance
labels = dbscan.labels_

# Print the number of clusters found
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print('Number of clusters found: ', n_clusters)

# Print the indices of core instances
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
print('Indices of core instances: ', np.where(core_samples_mask))

# Print the indices of border instances
border_samples_mask = np.logical_xor(core_samples_mask, labels == -1)
print('Indices of border instances: ', np.where(border_samples_mask))

# Print the indices of noise instances
noise_samples_mask = labels == -1
print('Indices of noise instances: ', np.where(noise_samples_mask))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN

# Generate data
X = np.random.randn(1000, 2)

# Make a copy of data and add a constant value
X_copy1 = np.concatenate([X, X + 5], axis=0)

# Make a copy of data and subtract a constant value
X_copy2 = np.concatenate([X, X - 5], axis=0)

# Merge all datasets
X_merged = np.concatenate([X, X_copy1, X_copy2], axis=0)

# Fit DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=10)
dbscan.fit(X_merged)

# Get cluster labels and number of clusters
labels = dbscan.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

# Plot the clusters
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(labels))))
for label, color in zip(set(labels), colors):
    if label == -1:
        # Black points are noise points
        color = 'k'
    class_member_mask = (labels == label)
    xy = X_merged[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], c=color, s=20, edgecolor='k')

plt.title('Estimated number of clusters: %d' % n_clusters)
plt.show()
```
